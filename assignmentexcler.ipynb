```python
#Day-3
#DAY-3Take input for marks in three subjects
# Take input for marks in three subjects
subject1 = float(input("Enter marks for Subject 1: "))
subject2 = float(input("Enter marks for Subject 2: "))
subject3 = float(input("Enter marks for Subject 3: "))

# Calculate the average
average = (subject1 + subject2 + subject3) / 3

# Determine the grade based on the average
if average >= 90:
    print("Grade: A")
elif 80 <= average < 90:
    print("Grade: B")
elif 70 <= average < 80:
    print("Grade: C")
else:
    print("Grade: Fail")
```

```python
#Day-4
# Take input for the positive integer n
n = int(input("Enter a positive integer: "))

# Initialize a variable to store the sum of even numbers
even_sum = 0

# Loop through all numbers from 1 to n
for num in range(1, n + 1):
    if num % 2 == 0:  # Check if the number is even
        even_sum += num

# Print the result
print(f"The sum of all even numbers between 1 and {n} is: {even_sum}")
```

```python
#Day-6
import pandas as pd

# 1. Create a DataFrame
data = {
    "Name": ["John", "Alice", "Bob", "Diana"],
    "Age": [28, 34, 23, 29],
    "Department": ["HR", "IT", "Marketing", "Finance"],
    "Salary": [45000, 60000, 35000, 50000]
}
df = pd.DataFrame(data)

# 2a. Display the first 2 rows of the DataFrame
print("First 2 rows of the DataFrame:")
print(df.head(2))

# 2b. Add a new column named 'Bonus' (10% of the salary)
df["Bonus"] = df["Salary"] * 0.1
print("\nDataFrame with Bonus column:")
print(df)

# 2c. Calculate the average salary
average_salary = df["Salary"].mean()
print("\nAverage Salary:", average_salary)

# 2d. Filter and display employees older than 25
filtered_employees = df[df["Age"] > 25]
print("\nEmployees older than 25:")
print(filtered_employees)
```

```python
import pandas as pd

# Load the dataset
file_path = "Day_7_sales_data.csv"  # Replace with the correct path to your CSV file
sales_data = pd.read_csv(file_path)

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(sales_data.head())

# Print basic statistics of numerical columns
print("\nBasic statistics of numerical columns:")
print(sales_data.describe())

# Calculate the total sales for each region
total_sales_by_region = sales_data.groupby("Region")["Sales"].sum()
print("\nTotal sales for each region:")
print(total_sales_by_region)

# Find the most sold product (based on quantity)
most_sold_product = sales_data.groupby("Product")["Quantity"].sum().idxmax()
print("\nMost sold product (based on quantity):", most_sold_product)

# Compute the average profit margin for each product
sales_data["Profit Margin (%)"] = (sales_data["Profit"] / sales_data["Sales"]) * 100
average_profit_margin = sales_data.groupby("Product")["Profit Margin (%)"].mean()
print("\nAverage profit margin for each product:")
print(average_profit_margin)
```

```python
import pandas as pd

# Load the CSV file (use your file path here)
df = pd.read_csv('Day_8_sales_data.csv')

# Data Filtering
# Extract rows where sales are greater than 1000
sales_greater_than_1000 = df[df['Sales'] > 1000]
print("Rows where sales are greater than 1000:")
print(sales_greater_than_1000)

# Find all sales records for the 'East' region
east_sales = df[df['Region'] == 'East']
print("\nSales records for 'East' region:")
print(east_sales)

# Data Processing
# Add a new column 'Profit_Per_Unit' calculated as Profit / Quantity
df['Profit_Per_Unit'] = df['Profit'] / df['Quantity']

# Create another column 'High_Sales' where Sales > 1000 is labeled as 'Yes', else 'No'
df['High_Sales'] = df['Sales'].apply(lambda x: 'Yes' if x > 1000 else 'No')

# Display the updated DataFrame
print("\nUpdated DataFrame with new columns:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Day_11_banking_data.csv')

# Strip any extra spaces from column names
df.columns = df.columns.str.strip()

# Task 1: Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_transaction_by_account = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Create a bar plot
plt.figure(figsize=(8, 6))
total_transaction_by_account.plot(kind='bar', color='skyblue')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 2: Create a pie chart to show the percentage of transactions per Branch
transactions_per_branch = df['Branch'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
transactions_per_branch.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # Hide the y-label as it's not needed for a pie chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv('Day_13_Pharma_data.csv')  # Replace 'your_dataset.csv' with the actual file name

# Data Cleaning: Check for missing values and duplicates
print("Missing values in the dataset:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Visualization 1: Bar plot comparing average Effectiveness for each drug across different regions
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Effectiveness', hue='Product_Name', data=df, ci=None)
plt.title('Average Effectiveness of Drugs by Region')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.show()

# Visualization 2: Violin plot to show the distribution of Effectiveness and Side_Effects for each product
plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df, inner="quart")
plt.title('Distribution of Effectiveness for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Effectiveness')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Side_Effects', data=df, inner="quart")
plt.title('Distribution of Side Effects for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Side Effects')
plt.show()

# Visualization 3: Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.suptitle('Pairplot: Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()

# Visualization 4: Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(8, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness Across Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()

# Visualization 5: Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Effectiveness vs Marketing Spend')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
